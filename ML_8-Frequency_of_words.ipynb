# Problem Statement

#### In this assignment students have to find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

#### Import necessary libraries

```python
from bs4 import BeautifulSoup 
import urllib.request 
import nltk 
from nltk.corpus import stopwords 
%matplotlib inline
```

#### get text from url and keep text only

```python
response = urllib.request.urlopen('http://php.net/') 
html = response.read() 
soup = BeautifulSoup(html,"html5lib") 
text = soup.get_text(strip=True) #strip=TRue to remove unwanted spaces
```

```python
print(text)
```

#### Examine frequency distribution of words

```python
tokens = [t for t in text.split()]
freq=nltk.FreqDist(tokens)
```

```python
freq.plot(20,cumulative=False)
```

Note: There are multiple stop words with high frequency which must be removed before proceeding further

#### Clean Stop words like 'this','of', 'and' etc

```python
clean_tokens = tokens[:] 
sr = stopwords.words('english') 
for token in tokens: 
    if token in stopwords.words('english'): 
        clean_tokens.remove(token) 
```

```python
clean_tokens
```

```python
freq = nltk.FreqDist(clean_tokens)
freq.plot(20,cumulative=False)
```

#### View the frequency of words accross on all sentences

```python
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```
